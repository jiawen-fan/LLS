```python
def unPackRawFile(raw_path):
    """


    
    - unpacks the .raw file. Not used for the neural network.
    """
    y = np.loadtxt(raw_path, skiprows=0)
    distance = y[:,0]
    pec_vel = y[:,1]
   
    return distance, pec_vel
```

```python
eigen_los,counts = unPackRawFile("investigate.raw")
counts, original = unPackRawFile("extract.raw")
counts, angle = unPackRawFile("angles.raw")
```

```python
def unPackRawFile2(raw_path):
    """


    
    - unpacks the .raw file. Not used for the neural network.
    """
    y = np.loadtxt(raw_path, skiprows=6)
    distance = y[:,0]
    pec_vel = y[:,1]
    temp = y[:,2]
    HI_density = y[:,3]
    gas_density = y[:,4]
    gas_metallicity = y[:,5]
   
    return distance, pec_vel, temp, HI_density, gas_density, gas_metallicity
```

```python
for i in range(0,51):
    raw_path = 'los.00' + '{0:03}'.format(i) +'.raw'
    distance, pec_vel, temp, HI_density, gas_density, gas_metallicity = unPackRawFile2(raw_path)
    plt.xlim(1.95,2.05)
    plt.plot(distance,HI_density)
    plt.show()
```
