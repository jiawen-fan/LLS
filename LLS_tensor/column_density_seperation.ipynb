```python
import yt
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import pickle as pickle
from unyt import g,cm
from numpy.linalg import eig

with open('angle_max_eigen.pkl','rb')as f:
    angle = pickle.load(f)

with open('lyman_cul.pkl','rb')as f:
    lyman = pickle.load(f)

plt.scatter(lyman,angle,s=0.1)
plt.xscale("log")
```

```python
data =[]
for i,k in zip(lyman, angle):
    data.append([i,k])
```

```python
data.sort()
```

```python
data
```

```python
bins = []
min = 1.6022738131189712e+16; max = 6.04643739275747e+21
bins = 10**np.linspace(np.log10(min),np.log10(max),6)
```

```python
bins
```

```python
_16_17 = []
for i in data:
    if i[0]<bins[1]:
        _16_17.append(i[1])
_17_18 = []
for i in data:
    if bins[1]<i[0]<bins[2]:
        _17_18.append(i[1])
_18_19 = []
for i in data:
    if bins[2]<i[0]<bins[3]:
        _18_19.append(i[1])
_19_20 = []
for i in data:
    if bins[3]<i[0]<bins[4]:
        _19_20.append(i[1])
_20_21 = []
for i in data:
    if bins[4]<i[0]<bins[5]:
        _20_21.append(i[1])
```

```python
var = plt.hist(_16_17,bins = 7)
```

```python
var1 = plt.hist(_17_18,bins = 7)
```

```python
var2 = plt.hist(_18_19,bins = 7)
```

```python
var3 = plt.hist(_19_20,bins = 7 )
```

```python
var4 = plt.hist(_20_21,bins = 7)
```

```python
normalized  = var[0]/np.sum(var[0])
normalized1  = var1[0]/np.sum(var1[0])
normalized2  = var2[0]/np.sum(var2[0])
normalized3  = var3[0]/np.sum(var3[0])
normalized4  = var4[0]/np.sum(var4[0])
```
