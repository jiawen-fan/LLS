```python
import yt
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import pickle as pickle
from unyt import g,cm
from numpy.linalg import eig

with open('angle_max_eigen.pkl','rb')as f:
    angle = pickle.load(f)

with open('lyman_cul.pkl','rb')as f:
    lyman = pickle.load(f)
```

```python
data =[]
for i,k in zip(lyman, angle):
    data.append([i,k])
```

```python
data.sort()
```

```python
len(data)
```

```python
upper = data[1448:]
```

```python
upper_data = []
for i in upper:
        upper_data.append(i[1])
```

```python
lower = data[:724]
```

```python
lower_data = []
for i in lower:
        lower_data.append(i[1])
```

```python
middle = data[724:1448]
```

```python
middle_data = []
for i in middle:
        middle_data.append(i[1])
```

```python
var,x,_ = plt.hist(lower_data,bins = 7)
normalized = var/sum(var)
var1,x,_ = plt.hist(upper_data,bins = 7)
normalized1 = var1/sum(var1)
total,x,_ = plt.hist(angle,bins = 7)
normalized2 = total/sum(total)
var2,x,_ = plt.hist(middle_data,bins = 7)
normalized3 = var2/sum(var2)
```

```python
bin_centers = 0.5*(x[1:]+x[:-1])
```

```python
plt.plot(bin_centers,normalized,c ="black",label = "lower")
plt.plot(bin_centers,normalized1,c ="green",label = "upper")
plt.plot(bin_centers,normalized2,c ="red",label = "everything")
plt.plot(bin_centers,normalized3,c ="blue",label = "middle")
plt.legend()
plt.ylim(0.05,.25)
plt.xlim(0,1)
```

```python
plt.plot(bin_centers,normalized2,c ="black",label = "All")
plt.fill_between(bin_centers,normalized2,normalized,color ="green",label = "Lower Quartile", alpha=0.5)
plt.fill_between(bin_centers,normalized2,normalized1,color ="red",label = "Upper Quartile", alpha=0.5)
plt.legend()
plt.ylim(0.05,.25)
plt.xlim(0,1)
```
