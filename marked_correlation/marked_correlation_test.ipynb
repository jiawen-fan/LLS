```python
import pickle
import numpy as np
from matplotlib import pyplot as plt  
import halotools.mock_observables as mo    
```

```python
Npts = 1000
Lbox = 1.0
period = np.array([Lbox,Lbox,Lbox])
```

```python
x = np.random.random(Npts)
y = np.random.random(Npts)
z = np.random.random(Npts)
```

```python
x
```

```python
coords = np.vstack((x,y,z)).T
```

```python
coords
```

```python
marks = np.random.random(Npts)
```

```python
rbins = np.logspace(-2,-1,10)
MCF = mo.marked_tpcf(coords,rbins, marks1=marks, period=period, normalize_by='number_counts', weight_func_id=1)
```

```python
MCF
```

```python
import matplotlib.pyplot as plt
plt.loglog(rbins[1:], MCF, label='Marked value')
plt.xlabel('r [Mpc/h]')
plt.ylabel('Marked value')
plt.legend()
plt.show()
```
