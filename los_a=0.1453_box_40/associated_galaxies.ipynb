```python
with open('a_lyman_limited_value.pkl', 'rb') as f:
    lyman = pickle.load(f)
```

```python
def associated_sphere(all_distance,data,counts):
    """
    this function finds every galaxy associated with lls in a sphere radius.
    """
    ##going through every los
    distance_from_lls = []
    lyman = []
    galaxies = []
    for i in range(counts):
        ##lyman
        for k in all_distance[i]:
            ##galaxies
            for z in range(len(data[i][1])):
                ## lyman direction from galaxies
                #might need to double check distance convertion(distance away)
                if(np.sqrt((k[3]-data[i][1][z])**2 + (data[i][2][z]/0.1453*0.69)**2) <.15):
                    ##lyman is the all_distance, must includes column density, left,right, middle, peak information, and the last for position
                    ##data is the galaxies information, this includes, halo, postion.
                    lyman.append([k[4][0],k[4][1],k[4][2]])
                    galaxies.append([data[i][0][z],data[i][3][z],data[i][4][z],data[i][5][z],np.sqrt((k[3]-data[i][1][z])**2 + (data[i][2][z]/0.1453*0.69)**2)])
                    distance_from_lls.append([data[i][3][z]-k[4][0],data[i][4][z]-k[4][1],data[i][5][z]-k[4][2]])
                    ##this returns, distance difference in projection, halo size, galaxies location, xyz respectively, lyman limited, xyz respectively,
    return lyman,galaxies,distance_from_lls
```

```python
#gives halo, distance along LOS, and distance away from LOS,x,y,z
#column density, left base,right base, middle, (x,y,z))
```

```python
lyman_to_append,galaxies,distance_from_lls = associated_sphere(lyman,data,30)
stuff = []
stuff.append(lyman_to_append)
stuff.append(galaxies)
stuff.append(distance_from_lls)
```
