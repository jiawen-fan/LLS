```python
import yt
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import pickle as pickle
from unyt import g,cm
from numpy.linalg import eig
```

```python
with open("angle_3.pkl",'rb') as f:
    angle= pickle.load(f)
```

```python
normalized  = var[0]/np.sum(var[0])
```

```python
x = np.linspace(0,1,10)
```
