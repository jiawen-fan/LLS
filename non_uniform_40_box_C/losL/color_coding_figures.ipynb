```python
def color_coding(i):
    if(i == 1):
        return str("blue")
    if(i == 2):
        return str("red")
    if(i == 3):
        return str("yellow")
    if(i == 4):
        return str("green")
    if(i == 5):
        return str("black")
```

```python
color = []
for i in lymans:
    color.append(color_coding(i[1]))
```
